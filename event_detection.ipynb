```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
 

path_gothic = 'C:\\Windows\\Fonts\\malgunsl.ttf'
fontprop1 = fm.FontProperties(fname=path_gothic, size=40)
```

```python
xls['현재지수']=xls['현재지수'].str.replace(',', '')
list_x=xls['현재지수'].tolist()
list_x=list(map(float,list_x))

import warnings
warnings.filterwarnings('ignore')

plt.rc('font', family='NanumGothic')
plt.figure(figsize=(20,20))     # 그래프 크기(단위 inch)
plt.xlabel('일자',fontproperties=fontprop1)
plt.ylabel('코스피 지수',fontproperties=fontprop1)
plt.title('일자별 코스피 지수',fontproperties=fontprop1)
plt.xticks(rotation=60)



plt.plot(xls['일자'], list_x,'r')
```
