# 크롤링(crawling) 이란?

Web상에 존재하는 Contents를 수집하는 작업
1. HTML 페이지를 가져와서, HTML/CSS등을 파싱하고, 필요한 데이터만 추출하는 기법
2. Open API(Rest API)를 제공하는 서비스에 Open API를 호출해서, 받은 데이터 중 필요한 데이터만 추출하는 기법
3. Selenium등 브라우저를 프로그래밍으로 조작해서, 필요한 데이터만 추출하는 기법

BeautifulSoup 라이브러리
- HTML의 태그를 파싱해서 필요한 데이터만 추출하는 함수를 제공하는 라이브러리

find() : 가장 먼저 검색되는 태그 반환
find_all() : 전체 태그 반환


```python
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get('http://v.media.daum.net/v/20170615203441266')
```

```python
print(res.content)
```

```python
soup = BeautifulSoup(res.content, 'html.parser')
```

```python
type(soup)
```

```python
soup
```

```python
soup.body
```

```python
type(soup.body)
```

```python
div = soup.body.find_all('div', class_='newsview_media')
div
```

```python
type(div)
```

```python
div.find_all('div', role='main')
```

```python
div_1 = soup.body.find_all('div', class_='newsview_media')[0]
div_1
```

```python
type(div_1)
```

```python
div_2 = div_1.find_all('div', role='main')[0]
div_2
```

```python
div_3 = div_2.find_all('div', id="cMain")[0]
div_3 
```

```python
div_4 = div_3.find_all('div', class_="foot_view")[0]
div_4
```

```python
div_5 = div_4.find_all('div', class_ = "cp_view")[0]
div_5
```

```python
news_list = div_5.find_all("li")
news_list
```

```python
text_sum = []
for item in news_list:
    text_sum.append(item.text)
```

```python
text_sum
```

```python
final_url = []
for item in news_list:
    final_url.append(item.a["href"])
```

```python
final_url
```

연습: http://media.daum.net/digital 에서 뉴스 리스트와 URL을 불러와보세요.

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

html = urlopen("http://www.naver.com")  

bsObject = BeautifulSoup(html, "html.parser") 

print(bsObject)
```

```python
print(bsObject.head.title)
```

```python
for meta in bsObject.head.find_all('meta'):
    print(meta.get('content'))
```

```python
print (bsObject.head.find("meta", {"name":"description"}).get('content'))
```

```python
for link in bsObject.find_all('a'):
    print(link.text.strip(), link.get('href'))
```
